# Location Analytics Dashboard

This notebook uses outputs from `outputs/location_analytics` and dashboard assets from `outputs/location_analytics/dashboard`.


```python
from pathlib import Path
import pandas as pd

BASE = Path('outputs/location_analytics')
DASH = BASE / 'dashboard'

freq = pd.read_csv(BASE / 'location_frequency_by_period.csv')
overall = pd.read_csv(BASE / 'location_frequency_overall.csv')
growth = pd.read_csv(BASE / 'location_growth.csv')
cooc = pd.read_csv(BASE / 'location_cooccurrence.csv')
sent = pd.read_csv(BASE / 'location_sentiment.csv')

periods = sorted(freq['period'].dropna().unique().tolist())
period_a, period_b = periods[0], periods[1]
print('Period A:', period_a)
print('Period B:', period_b)
print('Rows:', {
    'freq': len(freq),
    'overall': len(overall),
    'growth': len(growth),
    'cooc': len(cooc),
    'sent': len(sent),
})
```

## Quick Tables

```python
display(overall.head(15))
display(growth.sort_values('mention_delta', ascending=False).head(15))
display(cooc.sort_values('co_mentions', ascending=False).head(15))
```

## Dashboard Charts and Maps

```python
from IPython.display import SVG, display, Markdown

assets = [
    'chart_top_locations_overall.svg',
    'chart_top_locations_by_period.svg',
    'chart_growth_top_delta.svg',
    'chart_sentiment_top_locations.svg',
    'map_mentions_period_a.svg',
    'map_mentions_period_b.svg',
    'map_growth_delta.svg',
]

for name in assets:
    path = DASH / name
    display(Markdown(f'### `{name}`'))
    if path.exists():
        display(SVG(filename=str(path)))
    else:
        print('Missing:', path)
```

## Filter by Location

```python
location = 'Dhaka'  # change this

print('Frequency by period')
display(freq[freq['location'] == location].sort_values('period'))

print('Sentiment split')
display(sent[sent['location'] == location].sort_values(['period', 'sentiment']))

print('Growth row')
display(growth[growth['location'] == location])
```
